```python
import pymongo
from pymongo import MongoClient
from pymongo import MongoClient
client = MongoClient('localhost', 27017)
db = client.appdata
import datetime
import pprint
import pandas as pd
import numpy as np
```

```python
connection= db.androidapps
```

```python
ls=[]
ID=[]

for con in connection.find({'subcategory_name':"Sports"}):
    #pprint.pprint(con)
    con["App"]="android"
    ID.append(con["id"])
    ls.append(con)
df=pd.DataFrame(ls)
df
```

```python
connection2= db.iosapps
for con in connection2.find({'subcategory_name':"Sports"}):
    #pprint.pprint(con)
    con["App"]="ios"
    con['id']=str(con["id"])
    
    ID.append(str(con["id"]))
    ls.append(con)
df=pd.DataFrame(ls)
```

```python
df.to_csv("sports_meta.csv")
```
